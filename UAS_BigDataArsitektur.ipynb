```python
import pandas as pd

# Load the dataset
file_path = '/content/YOUTUBE CHANNELS DATASET.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Check data types and missing values
data.info()

# Handle missing values if necessary
data.fillna(0, inplace=True)  # Replace NaN with 0 for simplicity
```

```python
# Remove commas and convert to numeric
data['Subscribers'] = data['Subscribers'].str.replace('M', '').astype(float) * 1e6
data['Uploads'] = pd.to_numeric(data['Uploads'], errors='coerce').fillna(0)
data['Views'] = data['Views'].str.replace(',', '').astype(float)
```

```python
import pygal

# Get top 10 channels by subscribers
top_channels = data.nlargest(10, 'Subscribers')

# Create a bar chart
bar_chart = pygal.Bar()
bar_chart.title = 'Top 10 YouTube Channels by Subscribers'
for index, row in top_channels.iterrows():
    bar_chart.add(row['Username'], row['Subscribers'])

# Save the chart to a file
bar_chart.render_to_file('/content/top_youtube_channels.svg')
```

```python
import plotly.express as px

# Group by country and sum subscribers
country_data = data.groupby('Country')['Subscribers'].sum().reset_index()

# Create a pie chart
fig = px.pie(country_data, values='Subscribers', names='Country', title='Subscribers Distribution by Country')
fig.show()
```
