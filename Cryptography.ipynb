```python
import cv2
import numpy as np
from cryptography.fernet import Fernet
from google.colab.patches import cv2_imshow

# Step 1: Generate and save encryption key
def generate_key():
    key = Fernet.generate_key()
    with open("encryption_key.key", "wb") as key_file:
        key_file.write(key)
    print("Encryption key generated and saved to 'encryption_key.key'.")

# Step 2: Load the encryption key
def load_key():
    with open("encryption_key.key", "rb") as key_file:
        return key_file.read()

# Step 3: Encrypt an image
def encrypt_image(image_path="Szoboszlai.jpg", output_path="encrypted_image.enc"):
    key = load_key()
    cipher = Fernet(key)

    # Read image as binary data
    with open(image_path, "rb") as image_file:
        image_data = image_file.read()

    # Encrypt the binary data
    encrypted_data = cipher.encrypt(image_data)

    # Save the encrypted data to a file
    with open(output_path, "wb") as encrypted_file:
        encrypted_file.write(encrypted_data)
    print(f"Image encrypted and saved as '{output_path}'.")

# Step 4: Decrypt an image
def decrypt_image(encrypted_path="encrypted_image.enc", output_path="decrypted_Szoboszlai.jpg"):
    key = load_key()
    cipher = Fernet(key)

    # Read encrypted data
    with open(encrypted_path, "rb") as encrypted_file:
        encrypted_data = encrypted_file.read()

    # Decrypt the binary data
    decrypted_data = cipher.decrypt(encrypted_data)

    # Save the decrypted data as an image
    with open(output_path, "wb") as image_file:
        image_file.write(decrypted_data)
    print(f"Image decrypted and saved as '{output_path}'.")

# Step 5: Demonstration
def main():
    print("Image Encryption and Decryption Demo")
    print("1. Generate Key")
    print("2. Encrypt Image")
    print("3. Decrypt Image")
    print("4. Exit")

    while True:
        choice = input("Select an option (1-4): ")
        if choice == "1":
            generate_key()
        elif choice == "2":
            encrypt_image()
        elif choice == "3":
            decrypt_image()
            decrypted_image = cv2.imread("decrypted_Szoboszlai.jpg")
            print("Decrypted Image:")
            cv2_imshow(decrypted_image)
        elif choice == "4":
            print("Exiting application.")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
