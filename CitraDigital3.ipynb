```python
from PIL import Image
import matplotlib.pyplot as plt
import os
from google.colab import files

# Function to upload and display image
def upload_image():
    uploaded = files.upload()
    image_name = list(uploaded.keys())[0]
    image = Image.open(image_name)
    return image, image_name

# Function to resize image and display before and after
def resize_image(image, new_size):
    # Original image size
    original_size = image.size

    # Resize image
    resized_image = image.resize(new_size, Image.Resampling.LANCZOS)

    # Display original and resized images
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    ax[0].imshow(image)
    ax[0].set_title(f'Original Image\nSize: {original_size}')
    ax[0].axis('off')

    ax[1].imshow(resized_image)
    ax[1].set_title(f'Resized Image\nSize: {new_size}')
    ax[1].axis('off')

    plt.show()
    return resized_image

# Main workflow
print("Upload a 4K image (3840 x 2160):")
image_4k, image_4k_name = upload_image()

# Task 1: Reduce 4K image to 128 x 128
print("\nReducing the 4K image to 128 x 128:")
resized_4k_to_128 = resize_image(image_4k, (128, 128))

# Save the resized image
resized_4k_to_128.save(f'resized_{os.path.splitext(image_4k_name)[0]}_128x128.png')

# Task 2: Upload a 128 x 128 image and resize it to 4K
print("\nUpload a 128 x 128 image:")
image_128, image_128_name = upload_image()

# Task 2: Increase 128 x 128 image to 4K (3840 x 2160)
print("\nIncreasing the 128 x 128 image to 4K (3840 x 2160):")
resized_128_to_4k = resize_image(image_128, (3840, 2160))

# Save the resized image
resized_128_to_4k.save(f'resized_{os.path.splitext(image_128_name)[0]}_4K.png')
```
